```python
import pandas as pd

# List file CSV yang mau digabung
files = ["/content/demoDPR.csv", "/content/demoDPRup.csv", "/content/demoDPRup1.csv"]

#@title Baca semua CSV dan gabungkan jadi 1 DataFrame
df = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

# Cek hasil
print("Jumlah total data:", len(df))
display(df.head())

# Simpan ke file gabungan
df.to_csv("demoDPR_gabungan.csv", index=False, encoding="utf-8-sig")
print("✅ Data berhasil digabung ke demoDPR_gabungan.csv")
```
